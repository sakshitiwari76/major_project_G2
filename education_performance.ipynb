**Step 1 : Import Libraries and Load Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Student_Performance.csv')
df
```

```python
df.tail()
```

**Step 2 : Data Overview and Missing Values**

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['study_hours'].fillna(df['study_hours'].mean(), inplace=True)
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

**Remove Duplicates**

```python
df.drop_duplicates(inplace=True)
```

**Text data standardize**

```python
df['gender'] = df['gender'].str.lower().str.strip()
df['school_type'] = df['school_type'].str.lower()
df['internet_access'] = df['internet_access'].str.lower()
df['extra_activities'] = df['extra_activities'].str.lower()
df['final_grade'] = df['final_grade'].str.upper()
```

**Numeric range validation (Marks & Attendance)**

```python
# Attendance (0â€“100)
df = df[(df['attendance_percentage'] >= 0) &
        (df['attendance_percentage'] <= 100)]

# Subject marks (0â€“100)
score_cols = ['math_score', 'science_score', 'english_score']
for col in score_cols:
    df = df[(df[col] >= 0) & (df[col] <= 100)]
```

**Feature Engineering**

```python
df['avg_marks'] = df[['math_score','science_score','english_score']].mean(axis=1)
```

**Performance Level**

```python
df['performance_level'] = pd.cut(
    df['avg_marks'],
    bins=[0,40,60,80,100],
    labels=['Poor','Average','Good','Excellent']
)
```

**Step 3 : EDA (Exploratory Data Analysis)**

**1. Overall Score Distribution**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['avg_marks'], kde=True)
plt.title("Distribution of Average Marks")
plt.show()
```

**2. Subject-wise Average Performance**

```python
df[['math_score','science_score','english_score']].mean().plot(kind='bar')
plt.title("Subject-wise Average Scores")
plt.ylabel("Average Marks")
plt.show()
```

**3. Gender-wise Performance**

```python
sns.boxplot(x='gender', y='avg_marks', data=df)
plt.title("Gender-wise Academic Performance")
plt.show()
```

**4. Attendance vs Performance**

```python
sns.scatterplot(x='attendance_percentage', y='avg_marks', data=df)
plt.title("Attendance vs Academic Performance")
plt.show()
```

**5. Study Hours Impact**

```python
sns.scatterplot(x='study_hours', y='avg_marks', data=df)
plt.title("Study Hours vs Performance")
plt.show()
```

**6. Performance Level Count**

```python
df['performance_level'].value_counts().plot(kind='bar')
plt.title("Student Performance Levels")
plt.show()
```

**7. Correlation Heatmap**

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Features")
plt.show()
```

**Step 4 : Clean Data Export**
